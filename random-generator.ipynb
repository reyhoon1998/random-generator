```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Centeral limit theorem
N1 = int(input("N1 = "))
N2 = int(input("N2 = "))
```

```python
rand_s = np.random.rand(N2, N1)
```

```python
y = np.mean(rand_s, axis=1)
```

```python
plt.hist(y, bins=200)
plt.xlabel('number')
plt.ylabel('frequency')
plt.title('N2 = %s\nN1 = %s'%(N2, N1))
plt.show()
```

```python
#generating Gaussian with random number generator
N = int(input("N = "))
#desired sigma of our function
sigma = float(input('sigma = '))
#generating random numbers
x1 = np.random.rand(N)
x2 = np.random.rand(N)
#polar coordinate
r = np.sqrt(-2 * (sigma ** 2) * (np.log(1 - x1)))
t = 2 * math.pi * x2
#generating gaussian distribution
y1 = r * np.cos(t)
y2 = r * np.sin(t)
```

```python
plt.hist(y1, bins=500, label='y1')
plt.hist(y2, bins=500, label='y2')
plt.xlabel('number')
plt.ylabel('frequency')
plt.title('N = %s, $\sigma$ = %s' % (N, sigma))
plt.legend()
plt.show()
```
