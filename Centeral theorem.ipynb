```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

N1 = int(input("N1 = "))


N2 = int(input("N2 = "))
```

```python

rand_s = np.random.rand(N2, N1)
```

```python

y = np.mean(rand_s, axis=1)
```

```python
plt.hist(y, bins=200)
plt.xlabel('number')
plt.ylabel('frequency')
plt.title('N2 = %s\nN1 = %s'%(N2, N1))
plt.show()
```

```python
N1_s = np.array([5, 10, 100, 1000])
N2 = 1000000
```
